# Distribution of users rating

distribution of members ratinng and visulised in either bar chart and a donut chart

```python
rating_counts = movies['Movie Rating'].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.countplot(x=movies["Movie Rating"])

for i,rate in enumerate(rating_counts):
    plt.text(i,rate+2,f"{round(100*rate/movies.shape[0])}%",ha="center")
    
plt.subplot(1,2,2)   
movies["Movie Rating"].value_counts().plot.pie(autopct="%.0f%%",pctdistance=0.8,wedgeprops=dict(width=0.4))

plt.tight_layout()
plt.show()
```
